```python
#Load Packages
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = Path(r"C:\Users\lmeji\Desktop\Project_branch\Investing_Analysis\Resources\ExportingMainData_Gas_Stocks.csv")
data_df = pd.read_csv(data)
data_df.head()
```

```python
x = data_df['gspc_open'].values
y = data_df['weekly_regular_retail_gasoline_prices'].values
```

```python
x = x.reshape(-1, 1)
```

```python
#Degrees of Freedom = 2
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
```

```python
poly
```

```python
X_poly = poly.fit_transform(x.reshape(-1, 1))
```

```python
poly.fit(X_poly, y)
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(X_poly, y)
```

```python
y_pred = linreg.predict(X_poly)
y_pred
```

```python
plt.scatter(x,y, color='gray')
plt.plot(x, y_pred, color='red')
```

```python
plt.figure(figsize=(10, 6))
plt.title("Polynomial Regression df=2", size=16)
plt.scatter(x,y)
plt.plot(x, y_pred, color='red')
plt.show()
```
